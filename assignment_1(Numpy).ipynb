# 1.create a 5X5 array with numbers 1-25 , extract the middle 3X3 block .

```python
import numpy as np
```

```python
arr1=np.arange(1,26).reshape(5,5)
print(arr1)
```

```python
print(arr1[0:3])
```

# 2.create an array of10 numbers, replace all even numbers with -1.

```python
arr2=np.arange(1,11)
print(arr2)
arr2[arr2 % 2 == 0] = -1
print(arr2)
```
